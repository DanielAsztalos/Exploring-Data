```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
wine_data = pd.read_csv("https://raw.githubusercontent.com/vlewir/adatvizualizacio/main/winequality-red.csv")
wine_data.head()
```

```python
wine_data.describe()
```

```python
mean = wine_data.alcohol.mean()
mode = wine_data.alcohol.mode()[0]
median = wine_data.alcohol.median()
std = wine_data.alcohol.std()
```

```python
plt.figure(figsize=(10, 5))

sns.histplot(
    x="alcohol",
    data=wine_data,
    binwidth=0.2,
    edgecolor='white'
)

plt.axvline(mean, linestyle='--', color='r')
plt.axvline(mode, linestyle='--', color='g')
plt.axvline(median, linestyle='--', color='y')
plt.axvline(mean + std, linestyle='--', color='lightblue')
plt.axvline(mean - std, linestyle='--', color='darkblue')

plt.legend({"Mean": mean, "Mode": mode, "Median": median, "+1 Std": mean + std, "-1 Std": mean - std})

plt.title("Distribution of alcohol level in the wines")
plt.grid(axis='y', color='white', linewidth=0.25)
plt.xlabel("Alcohol")
```

```python
plt.figure(figsize=(7, 7))

sns.boxplot(
    x="quality",
    y="alcohol",
    data=wine_data
)

plt.axhline(median, color='lightblue', label='Global median')

plt.legend()

plt.title("Distribution of alcohol level in each quality category")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.plot()
```

```python
wine_data["scaled_alcohol"] = 0
wine_data
```

```python
medians = [0 for _ in wine_data.quality.unique()]

for q in wine_data.quality.unique():
    medians[q - wine_data.quality.unique().min()] = wine_data[wine_data.quality == q].alcohol.median()
    
for i in range(wine_data.shape[0]):
    wine_data.loc[i, "scaled_alcohol"] = wine_data.loc[i, "alcohol"] - medians[wine_data.loc[i, "quality"] - wine_data.quality.unique().min()]
    
wine_data
```

```python
plt.figure(figsize=(7, 7))

sns.boxplot(
    x="quality",
    y="scaled_alcohol",
    data=wine_data
)

plt.axhline(0, color='lightblue', label='Global mean')

plt.legend()

plt.title("Distribution of alcohol level in each quality category centered by medians")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.plot()
```

```python
bmi_data = pd.read_csv("https://raw.githubusercontent.com/vlewir/adatvizualizacio/main/BMI_fit_kaggle.csv", index_col=0)
bmi_data.head()
```

```python
sns.lmplot(
    x="Magassag (m)",
    y="BMI",
    data=bmi_data,
    order=2,
)

plt.legend(["2nd order regression line"])

plt.title("Relationship between BMI and height")
plt.xlabel("Height")
plt.xlim((1.35, 2.05))
plt.plot()
```
